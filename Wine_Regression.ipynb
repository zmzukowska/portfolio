```python
X = wine_data[["density"]]
Y = wine_data[["alcohol"]]

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25,random_state = 123)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import statsmodels.api as sm
from statsmodels.graphics.api import abline_plot
from sklearn import linear_model, preprocessing
```

```python
model = linear_model.LinearRegression(normalize=True)
```

```python
model.fit(X_train,Y_train)
```

```python
print(model.score(X_train,Y_train))
```

```python
Y_pred = model.predict(X_test)
```

```python
plt.scatter(Y_test,Y_pred)
plt.plot([x for x in range(9,15)], [y for y in range(9,15)], color='red')
```

```python
X=wine_data.drop(["alcohol","quality","free.sulfur.dioxide"],axis=1)
X=sm.add_constant(X)
Y=wine_data[["alcohol"]]
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=123)
```

```python
model3=sm.OLS(Y_train,X_train)
model3_results=model3.fit()
model3_results.summary()
```

```python
Y_pred=model3_results.predict(X_test)
```

```python
sns.regplot(Y_test,Y_pred)
plt.plot([x for x in range (9,15)],[y for y in range (9,15)], color='red')
```

```python
def rmse(predictions,targets):
    return np.sqrt(((predictions - targets)**2).mean())

Y_pred=model3_results.predict(X_test)

matches = pd.DataFrame(Y_test)
matches.rename(columns={'alcohol':'actual'}, inplace=True)
matches["predicted"] = Y_pred

matches.head()
```

```python
rmse(matches["predicted"],matches["actual"])
```

```python
plt.figure(figsize=(20,10))

sns.heatmap(matches, annot=True)
```
